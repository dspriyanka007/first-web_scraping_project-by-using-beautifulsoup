```python
import pandas as pd 
import numpy as np
import requests 
from bs4 import BeautifulSoup 
import pprint # “pprint” print data structures in well-formatted & more readable way
```

```python
res = requests.get("https://news.ycombinator.com/newest")
```

```python
print(res.text)
```

```python
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
print(soup.title)
```

```python
print(soup.find('a'))
```

```python
print(soup.find_all('div'))
```

```python
links = soup.select('.storylink')
subtext = soup.select('.subtext')
```

#### We create function for collecting title, link and vote of each hacker news in dict form

```python
def create_custom_hn(links, subtext):
    hn = []
    for idx, item in enumerate(links):
        title = links[idx].getText()
        href = links[idx].get('href', None)
        vote = subtext[idx].select('.score')
        if len(vote):
            points = (vote[0].getText())#.replace(' points' , ''))
            hn.append({'title': title, 'link': href, 'vote': points})
    return hn          
```

```python
print(create_custom_hn(links, subtext)) 
```

```python
pprint.pprint(create_custom_hn(links, subtext))
```

#### define a function for sort stories in hacker news by votes

```python
def sort_stories_by_votes(hnlist):
    return sorted(hnlist, key = lambda k:k['vote']) 
def create_custom_hn(links, subtext):
    hn = []
    for idx, item in enumerate(links):
        title = links[idx].getText()
        href = links[idx].get('href', None)
        vote = subtext[idx].select('.score')
        if len(vote):
            points = (vote[0].getText())#.replace(' points' , ''))
            hn.append({'title': title, 'link': href, 'vote': points})
    return sort_stories_by_votes(hn)  
```

```python
pprint.pprint(create_custom_hn(links, subtext))
```

#### for reverse sort order just add 'reverse = True' at very last in second line of define function of sort_stories_by_votes

```python
def sort_stories_by_votes(hnlist):
    return sorted(hnlist, key = lambda k:k['vote'], reverse =True) 
def create_custom_hn(links, subtext):
    hn = []
    for idx, item in enumerate(links):
        title = links[idx].getText()
        href = links[idx].get('href', None)
        vote = subtext[idx].select('.score')
        if len(vote):
            points = (vote[0].getText())#.replace(' points' , ''))
            hn.append({'title': title, 'link': href, 'vote': points})
    return sort_stories_by_votes(hn)
```

```python
pprint.pprint(create_custom_hn(links, subtext))
```
